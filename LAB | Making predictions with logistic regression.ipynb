```python
# prep: import modules and get pwd
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
# get the data
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)

```

```python

query = ''' SELECT 
    f.film_id,
    f.title,
    l.name AS 'language',
    f.rental_duration,
    f.rental_rate,
    f.length,
    f.rating,
    f.special_features,
    c.name AS 'category',
    r.rental_date,
    MONTH(r.rental_date) AS 'month'
FROM
    film f
        INNER JOIN
    inventory i ON f.film_id = i.film_id
        INNER JOIN
    language l ON f.language_id = l.language_id
        INNER JOIN
    film_category f_c ON f_c.film_id = f.film_id
        INNER JOIN
    category c ON f_c.category_id = c.category_id
        RIGHT JOIN
    rental r ON r.inventory_id = i.inventory_id; '''


df = pd.read_sql_query(query, engine)
print(df.shape)
df.head()
```

```python
list1 = []
for i in df['month']:
    if i == 5:
        list1.append(1)
    else:
        list1.append(0)

df['rented_may'] = list1

df['rented_may'].value_counts()
```

```python
df.head()
```

```python
# Drop columns rental_date and month to eliminate duplicates.


df2 = df.pivot_table(index= ['title','language','rental_duration','rental_rate','length','rating','special_features','category'], values = ['rented_may','film_id'], aggfunc = {'rented_may':max,'film_id':'count'}).reset_index()
```

```python
df2.rename(columns={"film_id": "count_rentals"}, inplace=True)
df2
```

### Train test split

```python
X_num = df2.select_dtypes(include = np.number)
X_cat = df2.select_dtypes(include = object)
y = df2['rented_may']
X = df2.drop('rented_may', axis=1)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1337)
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

X_train_num = X_train.select_dtypes(include = np.number)

# Scaling data
transformer = MinMaxScaler().fit(X_train_num) # need to keep transformer
X_train_normalized = transformer.transform(X_train_num)
X_train_norm = pd.DataFrame(X_train_normalized)
```

```python
X_train_norm.columns = X_train_num.columns
X_train_norm
```

```python
X_train_categorical = X_train.select_dtypes(include = object)
X_train_cat = pd.get_dummies(X_train_categorical, 
                             columns=['rating', 'category', 'special_features1','special_features2','special_features3','special_features4'],
                             drop_first=True)
X_train_categorical
```

```python
X_train_transformed = np.concatenate([X_train_norm, X_train_cat], axis=1)
```

### LOGISTIC REGRESSION MODEL

```python
from sklearn.linear_model import LogisticRegression
classification = LogisticRegression(random_state=0, solver='lbfgs',
                  multi_class='multinomial').fit(X_train_transformed, y_train)
```

```python
classification = LogisticRegression(random_state=0, solver='saga',
                  multi_class='multinomial').fit(X_train_transformed, y_train)
```

```python
# for numericals
X_test_num = X_test.select_dtypes(include = np.number)

# Scaling data
# we use the transformer that was trained on the training data
X_test_normalized = transformer.transform(X_test_num)
X_test_norm = pd.DataFrame(X_test_normalized)
```

```python
predictions = classification.predict(X_test_transformed)
classification.score(X_test_transformed, y_test)
```
